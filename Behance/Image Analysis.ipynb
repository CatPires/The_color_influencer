```python
## libraries
import glob
import re
import pandas as pd
```

```python
path_to_target = 'Behance\\'
path_to_file_list = glob.glob(path_to_target + '*.jpg')
```

```python
path_to_file_list
```

```python
images = [re.findall(r"(Behance\\)(\d{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^.]{0,})(.jpg)", image) for image in path_to_file_list]  ## get the image paths
#images ##print
```

```python
image_nr = [re.findall(r"(Behance\\)(\d{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^.]{0,})(.jpg)", image)[0][1] for image in path_to_file_list]  ## get the image paths
#image_nr
```

```python
likes = [re.findall(r"(Behance\\)(\d{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^.]{0,})(.jpg)", image)[0][3] for image in path_to_file_list]  ## get the image paths
#image_nr
```

```python
visualizations = [re.findall(r"(Behance\\)(\d{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^.]{0,})(.jpg)", image)[0][5] for image in path_to_file_list]  ## get the image paths
#image_nr
```

```python
date = [re.findall(r"(Behance\\)(\d{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^_]{0,})(_)([^.]{0,})(.jpg)", image)[0][9] for image in path_to_file_list]  ## get the image paths
#image_nr
```

```python
paths = [re.findall(r"(Behance\\)(\d{0,}_[^_]{0,}_[^_]{0,}_[^_]{0,}_[^.]{0,}.jpg)", image)[0][1] for image in path_to_file_list]  ## get the image paths
paths
#image_nr
```

```python
#lets retrieve dataframe with this information
projects_data = pd.DataFrame([path_to_file_list,image_nr,likes,visualizations,date,paths]).T
projects_data.columns = ['path','image_nr','likes','visualizations','date','simp_path']
```
